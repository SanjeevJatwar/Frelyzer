```python
import numpy as np
import pandas as pd

with open ("sample.txt","r") as file:
    text = file.read()
```

### Pre Processing

```python
# make all to lowercase and remove punctuation marks

text = text.lower()
```

```python
words  = text.split()
words
```

```python
from collections import Counter
word_freq = Counter(words)
```

### Result

```python
top_n = 10
for word, freq in word_freq.most_common(top_n):
    print(f"{word}: {freq}")
```
